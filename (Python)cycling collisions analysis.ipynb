```python
# Define a function to extract latitude and longitude
def extract_lat_long(geometry):
    coordinates = ast.literal_eval(geometry)['coordinates']
    latitude = coordinates[0][1]
    longitude = coordinates[0][0]
    return latitude, longitude

# Apply the function to create new columns for latitude and longitude
data2[['latitude', 'longitude']] = data2['geometry'].apply(lambda x: pd.Series(extract_lat_long(x)))
# Drop unnecessary column
data2.drop(columns=['geometry'], inplace=True)

# Print the DataFrame to verify the changes
print(data2.head())
```

```python
# Define a function to check if a collision happened on the cycling network
def collision_on_network(collision_row):
    for _, segment in data1.iterrows():
        if (segment['latitude1'] <= collision_row['latitude'] <= segment['latitude2'] or
            segment['latitude2'] <= collision_row['latitude'] <= segment['latitude1']) and (
            segment['longitude1'] <= collision_row['longitude'] <= segment['longitude2'] or
            segment['longitude2'] <= collision_row['longitude'] <= segment['longitude1']):
            return 'Yes'
    return 'No'

# Add a new column to data2 indicating if the collision happened on the cycling network
data2['on_cycling_network'] = data2.apply(collision_on_network, axis=1)

# Print the updated data2
print(data2.head())
```
